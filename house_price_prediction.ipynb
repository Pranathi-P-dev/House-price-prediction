```python
#Step-1 :Import all the Libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
#Step-2 : Create the data set 
data = {
    'Area(sq ft)' : [800,1000,1200,1500,1800],
    'Price(Lakhs)'  : [45,55,65,75,95]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Step-3 : Visualise the Data 
plt.scatter(df['Area(sq ft)'],df['Price(Lakhs)'],color = 'blue')
plt.xlabel('Area(sq ft)')
plt.ylabel('Price(Lakhs)')
plt.title('House Price vs Area')
plt.grid(True)
plt.show()
```

```python
# Prepare Data For Model
X = df[['Area(sq ft)']]
Y = df['Price(Lakhs)']

#Train Linear Regression Model 
model = LinearRegression()
model.fit(X,Y)

#Model Evaluation 
print(f"Model Coefficieant(slope): {model.coef_[0]}")
print(f"Model Intercept: {model.intercept_}")
print(f"R score: {model.score(X,Y)}")
```

```python
#Predictions 
area_input = np.array([[1300],[1600],[2000]])
predicted_prices = model.predict(area_input)

for area,price in zip(area_input,predicted_prices):
    print(f"predicted price for {area[0]}sq ft = {round(price, 2)}Lakhs")

#Plot Regression Line 
plt.scatter(X, Y,color = 'blue', label = 'Actual Data')
plt.plot(X, model.predict(X),color = 'red', label = 'Actual Data ')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price(Lakhs)')
plt.title('Linear Regression Model')
plt.legend()
plt.show()
```
